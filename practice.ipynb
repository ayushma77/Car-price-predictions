```python
import os
import pandas as pd 
import matplotlib.pyplot as plt
from PIL import Image 
```

```python
DATA_DIR="data"

folders= os.listdir(DATA_DIR)
print(folders)
```

```python
folders=[]
num_images=[]

# Loop through folder and count images
for folder in os.listdir(DATA_DIR):
    folder_path=os.path.join(DATA_DIR,folder)
    number_images=len(os.listdir(folder_path))
    folders.append(folder)
    num_images.append(number_images)
    
#creating a histogram
plt.figure(figsize=(10,6))
plt.bar(folders,num_images,color='skyblue')
plt.xlabel('Folders')
plt.ylabel('number of images')
plt.title("number of images in each folder")
plt.xticks(rotation=45,ha="right")
plt.tight_layout()
plt.show()
```

```python
#load and display image
#image_path=r"data\Mild\0a61bddab956.png"
image_path=os.path.join(DATA_DIR,"mild",os.listdir(os.path.join(DATA_DIR,"mild"))[0])
img=Image.open(image_path)
plt.imshow(img)
plt.show()
```
